## 2.1 <font color="green">Done</font> 

1. If we have a large set of variables, it becomes impossible to visualize and difficult to analyze it. We therefore might want to reduce the number of variables (dimensions), but that means that we will loose information. PCA is a method, that reduces dimensionality while keeping the loss of accuracy as small as possible.

2. There are exactly as many principal components as there are variables in the dataset, so in this case there are 5 principal components.

3. The principal components are ranked according to the eigenvalues of the covariance matrix, the bigger the eigenvalue, the more it is responsible for the variance
in the data. You can normalize the eigenvalues i.e. divide them by the sum of all eigenvalues to obtain the percentage of variance which that eigenvalue is responsible for.

4. If you apply the inverse transformation using only a subset of your principal components, you won't reproduce the original dataset perfectly. The reconstruction error is the difference between the original Dataset and the reduced one.

5. There should be no reconstruction error, since you don't reduce anything if you stay in the same number of (k) dimensions.

6. SVD can also be used for dimensionality reduction. Normally, PCA would do eigendecomposition on the covariance matrix, but many implementations seem to use SVD, since it is efficient and can handle sparse matrices.
