## 3.2 Special matrices (1 point)
Provide a definition and example for the following matrices:
1. Transpose of a matrix

The operation of transposing a matrix has a result the original matrix which has been flipped. This means that its orinal disposition of rows and columns has been changed so that the rows are now the column (and vice versa) of the resulting matrix.

E.g., $original matrix = \begin{bmatrix}3 & 4 & 7\\7 & 3 & 0\end{bmatrix}$

$transposed matrix = \begin{bmatrix}3 & 7\\4 & 3\\7 & 0\end{bmatrix}$

2. Identity matrix

An identity matrix has all entries on the diagonal from the upper left to the bottom right that are all equal to 1, and all other entries are 0.

E.g., $identity_matrix = \begin{bmatrix}1 & 0\\0 & 1\end{bmatrix}$

3. Inverse of a matrix

The inverse of a matrix A is denoted by $A^{-1}$. The original matrix needs to be a square matrix and its determinant needs to be different than zero. For a matrix A, its inverse A-1 = 1/|A| Adj A .

E.g., $original matrix = \begin{bmatrix}3 & 6\\-4 & 8\end{bmatrix}$


$inverse matrix = 1/48 · \begin{bmatrix}8 & -6\\4 & 3\end{bmatrix} = \begin{bmatrix}8/48 & -6/48\\4/48 & 3/48\end{bmatrix}$

4. Symmetric matrix

A square matrix is considered to be symmetrical if it is equal to if the matrix is transposed.

E.g., $original matrix = \begin{bmatrix}1 & -2\\-2 & 0\end{bmatrix}$


5. Orthogonal matrix

Square matrix whose transpose is equal to the inverse of the matrix.

E.g., $original matrix = \begin{bmatrix}1 & 0\\0 & 1\end{bmatrix}$
