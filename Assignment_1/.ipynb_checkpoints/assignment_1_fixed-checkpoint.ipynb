### Numpy
* Create Numpy array of size 30 filled with random integer numbers from 21 to 25
* Calculate mean and standard deviation of the array using Numpy functions.
* Write your own functions to calculate mean and std and validate that they are outputting the same values as Numpy functions.

### Pandas
* Create a Pandas DataFrame with two columns: 'Day' and 'Temperature'. 'Day' should have integer values from 1 to 30, and 'Temperature' is a Numpy array, generated at the previous step.
* Export resulting dataframe to csv file containing only 'Day' and 'Temperature' columns.

### Matplotlib
* Plot day-temperature graph as a simple 2D linear plot.
* Export resulting graph to png.

```python
# TODO

#### Numpy ####

a = np.random.randint(21,25, size = 30) # array with random integers
a_mean = np.mean(a)
a_std = np.std(a)

def get_mean(np_a):
    
    return (np_a.sum()/len(np_a))

    
def get_std(np_a):
    
    m = get_mean(np_a) # mean of array
    n = len(np_a) # number of elements in array
    
    dm = [] # data point -mean
    for i in np_a:
        dm.append((i-m)**2)
    dm = np.array(dm)
    dms = np.sum(dm)
    std_dev = np.sqrt(dms/(n))
    
    return std_dev

print ('Numpy\n')
print ('Array: ',a, '\n')
print ('Mean (numpy): ', a_mean)
print ('Mean (function): ', get_mean(a),'\n')
print ('Standard deviation (numpy): ', a_std)
print ('Standard deviation (function): ', get_std(a), '\n\n')


#### Pandas ####
days = np.arange(1,31,dtype=int)
temp = a
d = {'Days':days, 'Temperature':temp}
df = pd.DataFrame(data=d)

df.to_csv('DayTemp.csv')

#### Matplotlib ####
fig = plt.figure()
ax = plt.axes()

ax.plot(days,temp)
plt.savefig('DayTemp.png')
```
