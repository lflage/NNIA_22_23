In this exercise we will be working with the [MNIST](https://pytorch.org/vision/stable/generated/torchvision.datasets.MNIST.html#torchvision.datasets.MNIST) dataset, which consists consists of handwritten digits. Use the following script to run a 2 layer neural network on MNIST where cross entropy loss is used as the loss criterion. At each iteration of the training the accuracy of network on the test set is computed.  By default the network has the following structure $L_{28\cdot28\rightarrow 100} \circ \text{ReLU} \circ L_{100 \rightarrow 10}$, a learning rate of $0.1$ and a batch size of $200$.  
Experiment with 3 additional values for the learning rate, hidden size or batch size (choose only one) and run the training loop for at least 20 iterations (feel free to do more).

Plot the results for each value, including the default (4 plots in total). What do you observe?

```python
from mnist_2layer import run_nn
BATCH_SIZE = 200
EPOCH=40

# list of accuracies for default hyperparameters
accuracies_default = run_nn(batch_size=BATCH_SIZE, lr_rate=0.1, hidden_size=100, max_iter=EPOCH)
```

```python
# plot your results
# 

import matplotlib.pyplot as plt


to_plot = [i.item() for i in accuracies_default]
iters = list(range(1,EPOCH+1))
plt.figure(figsize=(15,6))
plt.plot(iters, to_plot)
plt.xlabel("Number of Training Epochs")
plt.xticks(iters)
plt.ylabel("Accuracy")
plt.title("Accuracy x Epoch - Default")
```

After one initial experiment we decided to increase the number of epochs to 40 in order to test if the accuracy could be improved further.
As we run the training in the new *default* setting for 40 epochs we found an accuracy at epoch 20 to be approximately 65.71, while at epoch 40 we achieve 75.52 with a max of 76.27 for epoch 38

In our next steps we decided to test how the Batch size influences the accuracy. Our first experiment ran with *batch size 50*, and, after a first look, we can see how unstable training becames, achieving 66.77 at the 18th iteration and later droping to 55.41 at the 22nd iteration. But the model eventually achieves accuracy similar to the default setting, 76.38 at the last training iteration.

For the next experiment we increased the batch size to 400. First thing to notice here is how the accuracy curve is smother, there is a lower variance in accuracy values for later epochs, when compared with batch size 50 and 200. Also, interesting to notice is the initial learning curve, which is steeper than in the last experiments. We can see that it hits an accuracy of 60% by the 13th iteration, while in the default setting it hits 60.36 at the 15th iteration.

This tendency of creating a steeper early curve is observed again in the last setting, with *batch size 600*, where the model hits 61.11 accuracy by the 9th iteration. This setting also obtains the highest final accuracy at 79.73

One last experiment with *batch size 1200* revealed that after a certain point the intial curve becomes more gentle, arriving at 60% accuracy by the 12th iteration, but hitting a final accuracy of 79.38.
